```python
response = ''
action = ''
users = []
user_map = {}

class Account:
    def __init__(self,name):
        self.balance = 0
        self.name = name
        print(f'Welcome {name} to our bank!')
        
    def deposit(self,amount):
        self.balance += int(amount)
        
    def withdraw(self,quantity):
        self.balance -= int(quantity)
        
    def checkBalance(self):
        return self.balance


def startApp():
    global response
    global name
    global users
    global action
    
    print('Welcome to the Bank!')
    while response not in ['new','existing']:
        response = input('Are you new or returning customer(new/existing)? ').lower()
        action = ''
        if response == 'new':
            #call function to show button to create new account
            create_new_account()
            while action != 'exit':
                actions()
            response = ''
        elif response == 'existing':
            name = input('Please enter your name: ').lower().capitalize()
            if name in user_map.keys():
                print(f'Welcome back {name}!')
                response = ''
                while action != 'exit':
                    actions()
            else:
                print('We couldn\'t find your record. Please try again!' )
                response = ''


def actions():
    global response
    global name
    global users
    global action
     #users['user'+str(len(users))] = Account(name)
    while action not in ['deposit', 'withdraw', 'check balance', 'exit']:
        action = input('What would you like to do(deposit/withdraw/check balance/exit): ')
        do_actions()
        
    return action

def get_user_balance(name):
    return print('Your current balance is ' + str(users[user_map[name]].checkBalance()))

def create_new_account():
    global name
    name = input('Please enter your full name: ').lower().capitalize() #capitalize first letter of the string
    #check if name is already in the list
    while name in user_map.keys():
        print('A user with this name already exists! Try again.')
        response = ''
        name = input('Please enter your full name: ').lower().capitalize()
    users.append(name)
    user_map[name] = users.index(name)
    users[users.index(name)] = Account(name)
    
    
def do_actions():
    global name
    global action
    if action.lower() == 'deposit':
        action = ''
        money = input('How much would you like to deposit? ')
        try:
            if int(money) > 0:
                users[user_map[name]].deposit(money)
                get_user_balance(name)
            else:
                print('Value must be an integer > 0!')
        except ValueError:
            money = ''
            print('Value must be an integer > 0!')
    elif action.lower() == 'check balance':
        action = ''
        get_user_balance(name)
    elif action.lower() == 'withdraw':
        action = ''
        if int(users[user_map[name]].checkBalance()) == 0:
            get_user_balance(name)
        elif int(users[user_map[name]].checkBalance()) > 0:
            quantity = input('How much would you like to withdraw? ')
            try:
                if int(quantity) > 0:
                    if int(quantity) <= int(users[user_map[name]].checkBalance()):
                        users[user_map[name]].withdraw(quantity)
                        get_user_balance(name)
                    else:
                        print('You do not have sufficient funds!')
                        get_user_balance(name)
                else:
                    print('Please enter a different amount!')
            except ValueError:
                print('Please enter a valid quantity!')
    elif action.lower() == 'exit':
#         action = ''
        print('Thank you for your business. Good bye!')
```

```python
startApp()
```
